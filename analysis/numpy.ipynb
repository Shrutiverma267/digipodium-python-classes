```python
t = np.arange(-np.pi*3, np.pi, .05)
x = np.sin(t)
y = np.cos(t)
print(t)
print(x)
print(y)
```

```python
import matplotlib.pyplot as plt
plt.plot(t,x,'r--o')
plt.plot(t,y,'g:x')
plt.title("Visualisation")
plt.show()
```
